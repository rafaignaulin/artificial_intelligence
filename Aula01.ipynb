# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python


plt.bar(df.bathrooms.value_counts().index, df.bathrooms.value_counts().values)

plt.title('Frequencia')
plt.xlabel('Banheiros')
plt.savefig('Banheiros.png')
plt.show()
```

```python
fig, ax = plt.subplots( nrows=1, ncols=2,figsize=(6,2))

ax[0].scatter(df['price'].value_counts().index, df['price'].value_counts())

ax[1].bar(df.bathrooms.value_counts().index, df.bathrooms.value_counts().values)
```
